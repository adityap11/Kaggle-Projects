```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('Data/RestaurantVisitors.csv')
```

```python
df.head()
```

```python
df.set_index('date', inplace=True)
df.index.freq = 'D'
```

```python
df.info()
```

```python
df1 = df.dropna()
```

```python
ax = df1['total'].plot(figsize=(16,8), title='Restaurant Visitors')
ax.autoscale(axis='x', tight=True)
ax.set(ylabel='Visitors per day');
```
